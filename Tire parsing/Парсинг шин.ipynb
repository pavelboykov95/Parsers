```python
from bs4 import BeautifulSoup
import requests
import lxml
import json
import csv
import time
import random
import os
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
import time

```

```python
cur_date = datetime.now().strftime("%d_%m_%Y_%H_%M_%S")

url = "https://roscarservis.ru/catalog/legkovye/?form_id=catalog_filter_form&filter_mode=params&sort=asc&filter_type=tires&filter%5C_mode=params&arCatalogFilter_458_1500340406=Y&set_filter=Y&isAjax=true&PAGEN_1=1"
headers = {
    "X-Is-Ajax-Request": "X-Is-Ajax-Request",
    "X-Requested-With": "XMLHttpRequest",
    "Accept": "application/json, text/javascript, */*; q=0.01",
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36"
}

r = requests.get(url=url, headers=headers)
if not os.path.exists("data8.1"):
    os.mkdir("data8.1")

pages_count = r.json()['pageCount']

data = []
stores_list = ["discountStores", "fortochkiStores", "commonStores"]
for page in range(1, 5):
#for page in range(1, pages_count + 1):
    url = f"https://roscarservis.ru/catalog/legkovye/?form_id=catalog_filter_\
    form&filter_mode=params&sort=asc&filter_type=tires&filter%5C_mode=params&arCatalogFilter_\
    458_1500340406=Y&set_filter=Y&isAjax=true&PAGEN_1={page}"
    
    r = requests.get(url=url, headers=headers)
    
    tires_data = r.json()['items']
    for tire_data in tires_data:
        tire_name = tire_data['name']
        tire_price = tire_data['price']
        image_url = f"https://roscarservis.ru{tire_data['imgSrc']}"
        url = f"https://roscarservis.ru{tire_data['url']}"
        total_amount = 0
        amount_data = []
        
        for store in stores_list:
            if store in tire_data:
                if tire_data[store] is None or len(tire_data[store]) < 1:
                    continue
                else:
                    for st in tire_data[store]:
                        store_amount = int(st['AMOUNT'])
                        print(store_amount)
                        total_amount+=store_amount
                        store_name = st['STORE_NAME']
                        store_price = st['PRICE']
                        amount_data.append(
                        {
                            "Наименование склада": store_name,
                            "Цена на складе": store_price,
                            "Кол-во на складе": store_amount
                        }
                        )
        data.append(
        {
            "Наименование": tire_name,
            "Цена": tire_price,
            "Общий остаток": total_amount,
            "Ссылка на изображение": image_url,
            "Ссылка на страницу": url,
            "Остатки на складах": amount_data
        }
        )
        
with open(f"Tire_data_8.1_{cur_date}.json", "a", encoding="utf-8") as file:
    json.dump(data, file, indent=4, ensure_ascii=False)
        
   
```
